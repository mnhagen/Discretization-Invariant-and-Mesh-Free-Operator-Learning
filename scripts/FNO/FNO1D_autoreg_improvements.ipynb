This file is meant to test different methods of improving the performance of the naively implemented autoregressive rollout by mitigating error accumulation.

# Methods to investigate:

- Progressive unrolling
    - change FNO loss calculations to apply the model multiple times and calculate loss once convergence is reached

- Multi-step teacher-forcing hybrid
    - gradually replace ground truth with model predictions during model training -> scheduled sampling

- striding? maybe teach the network to do multiple different time jumps and breaking the autoregressive jump into several larger ones. so instead of 0.715 = 143 * 0.005, do 0.715 = 0.5 + 2 * 0.1 + 3 * 0.005 or 0.715 = 7 * 0.1 + 3 * 0.005

- Multi-step loss
    - penalize errors over not just at the next step, but also over short unrolls

- Physics informed? ex. energy conservation. not sure if this will generalize well (will there always be applicable conservation laws?)
