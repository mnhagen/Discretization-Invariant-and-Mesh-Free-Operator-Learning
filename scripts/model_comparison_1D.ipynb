```python
import numpy as np
import matplotlib.pyplot as plt

CNO_path = "/scratch/mnhagen/models/heat1d/CNO1D_heat1d_N128_loss.csv"
CNO_stepLR_path = "/scratch/mnhagen/models/heat1d/CNO1D_heat1d_N128_stepLR_loss.csv"
FNO_path = "/scratch/mnhagen/models/heat1d/FNO1D_heat_eq_128_modes4_loss.csv"
FNO_stepLR_path = "/scratch/mnhagen/models/heat1d/FNO1D_heat_eq_128_modes4_stepLR_loss.csv"

CNO_arr = np.loadtxt(CNO_path, delimiter = ",")
CNO_stepLR_arr = np.loadtxt(CNO_stepLR_path, delimiter = ",")
FNO_arr = np.loadtxt(FNO_path, delimiter = ",")
FNO_stepLR_arr = np.loadtxt(FNO_stepLR_path, delimiter = ",")

CNO_time = CNO_arr[0, :]
CNO_loss = CNO_arr[1, :]
CNO_stepLR_time = CNO_stepLR_arr[0, :]
CNO_stepLR_loss = CNO_stepLR_arr[1, :]
FNO_time = FNO_arr[0, :]
FNO_loss = FNO_arr[1, :]
FNO_stepLR_time = FNO_stepLR_arr[0, :]
FNO_stepLR_loss = FNO_stepLR_arr[1, :]

plt.plot(CNO_time, CNO_loss, label = "CNO", c = "b")
plt.plot(FNO_time, FNO_loss, label = "FNO", c = "r")
plt.title("Loss as function of training time, scheduler = CosineAnnealingLR")
plt.yscale("log")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Loss (MSE)")
plt.show()


plt.plot(CNO_stepLR_time, CNO_stepLR_loss, label = "CNO", c = "b")
plt.plot(FNO_stepLR_time, FNO_stepLR_loss, label = "FNO", c = "r")
plt.title("Loss as function of training time, scheduler = stepLR")
plt.yscale("log")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Loss (MSE)")
plt.show()
```
